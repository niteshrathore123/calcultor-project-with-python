# Matplotlib library. 

### Matplotlib is an amazing visualization library in Python for 2D plots of arrays. Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack.


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]                     # x stands for the days number
y=[23,42,35,32,55,21,39]              #y stands for the temperature.
```

```python
plt.plot(x,y)                             #plt.plot() function is used for ploting the x,y coordinates. to check what arguments should take press---> shift+enter.
plt.show()
```

```python
plt.plot(x,y,'ro')                            
plt.show()                                     #ro mean ---> r stand for the red color and o stand for the dotted.           
```

```python
plt.plot(x,y,lw=4)                           #lw is used to change the width of line.                         
plt.show()                                               
```

### Now we have anaylyzing the data of the student score in their exams


```python
stu=[1,2,3,4,5,6]
first_exam=[90,78,67,85,81,73]                # this is the data
second_exam=[92,88,76,89,86,79]
```


     1. xlabel and ylabel is used to give label on the graphs.
     2. legend is used to  for better visualisation for the line label.

```python
plt.xlabel("Roll number")
plt.ylabel("Marks")
plt.plot(stu,first_exam,label="First midterm",lw=3)
plt.plot(stu,second_exam,label="second midterm",lw=3)
plt.legend()
plt.show()
```
